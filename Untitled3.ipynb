```python
import pandas as pd
import numpy as np
```

```python
data_team = pd.read_csv('teams_stats.csv', delimiter=',')
```

```python
data_team_new = data_team[['Year', 'Tm', 'Player', 'Age', 'Pos', 'PTS']]
data_team_new
```

```python
import math
data_team_new['Year'] = [int(0 if math.isnan(x) else x) for x in data_team_new['Year']]
data_team_new['Age'] = [int(0 if math.isnan(x) else x) for x in data_team_new['Age']]
data_team_new
```

```python
cnt = sum([int(1 if math.isnan(x) else 0) for x in data_team['Year']])
cnt
```

```python
data_new2 = data_team.copy()
positions = ['SG', 'PG', 'SF', 'PF', 'C']
data_new2['Pos'] = [p if p in positions else 'other' for p in data_new2['Pos']]
data_new2
```

```python
data_new2.groupby('Pos')['PTS'].mean().sort_values(ascending=False)
```

```python
data_new2.groupby(['Year', 'Pos']).mean().sort_values(by=['Year', 'PTS'], ascending=[True, False])
```
